```python
from nba_api.stats.endpoints import playbyplayv2
from collections import defaultdict
import pandas as pd

# Load play-by-play data
game_id = '0022200001'  # Replace with a valid NBA game ID
df = playbyplayv2.PlayByPlayV2(game_id=game_id).get_data_frames()[0]

# Set of all players mentioned in the game
players = set()
for col in ['HOMEDESCRIPTION', 'VISITORDESCRIPTION']:
    players.update(df[col].dropna().str.extractall(r'([A-Z][a-z]+\s[A-Z][a-z]+)')[0].values)

# Initialize player state
player_on_court = {p: False for p in players}
player_possessions = defaultdict(int)
player_success_possessions = defaultdict(int)

# TEMP: Set 5 players on each team at game start (replace with actual starters)
starting_players = list(players)[:10]
for p in starting_players:
    player_on_court[p] = True
print(f"🏀 Starters: {starting_players}")

# Step through each play
current_possession_team = None
for i, row in df.iterrows():
    desc = row['HOMEDESCRIPTION'] or row['VISITORDESCRIPTION'] or row['NEUTRALDESCRIPTION'] or ""
    event_type = row['EVENTMSGTYPE']
    
    # Handle substitutions
    if event_type == 8 and "enters the game" in desc:
        sub_in = desc.split(" enters the game")[0]
        for player in player_on_court:
            if player == sub_in:
                player_on_court[player] = True
                print(f"🔁 Sub In: {player}")
        continue
    elif event_type == 8 and "replaces" in desc:
        sub_out = desc.split(" replaces ")[1]
        for player in player_on_court:
            if player == sub_out:
                player_on_court[player] = False
                print(f"🔁 Sub Out: {player}")
        continue

    # Detect scoring play
    if event_type in [1, 3]:  # Made shot or FT
        scorer = None
        for player in player_on_court:
            if player in desc:
                scorer = player
                break
        
        for player in player_on_court:
            if player_on_court[player]:
                player_possessions[player] += 1
                if player != scorer:
                    player_success_possessions[player] += 1

        print(f"✅ Play: {desc.strip()}")
        for player in player_on_court:
            if player_on_court[player]:
                total = player_possessions[player]
                success = player_success_possessions[player]
                rate = (success / total) * 100 if total > 0 else 0
                print(f"   {player}: {rate:.2f}% IOC ({success}/{total})")
```
